```python
import sklearn
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from IPython.display import display
import seaborn as sns


X=pd.read_csv('train_preprocess.csv')
X_test=pd.read_csv('test_preprocess.csv')
y=pd.read_csv('train_target.csv',header=None)
y_train=y[[1]]
```

```python
from sklearn.preprocessing import StandardScaler
sc_train=StandardScaler()
sc_test=StandardScaler()
train_std=sc_train.fit_transform(X)
test_std=sc_test.fit_transform(X_test)

sc_y=StandardScaler()
y_std=sc_y.fit_transform(y_train)
```

```python
from keras.wrappers.scikit_learn import KerasRegressor
from keras.callbacks import EarlyStopping
from keras.models import Sequential
from keras.layers import Dense

def create_model():
    model=Sequential()
    model.add(Dense(16,input_dim=train_std.shape[1],activation='relu'))
    model.add(Dense(8,activation='relu'))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error',optimizer='Adam')
    return model

nn_reg=KerasRegressor(build_fn=create_model,epochs=100,callbacks=[EarlyStopping(monitor='val_loss',min_delta=0,patience=0,verbose=1,mode='auto')])
nn_reg.fit(train_std,y_std)
```

```python
import pickle
with open('dnn_0715.pickle','wb') as f:
    pickle.dump(nn_reg,f)
```
